```python
df[1] = df[1].drop(columns = ['reinforcement','reid_no', 'flags', 'accepted_rank', 'accepted_attr', 'ref_author', 'plant_organ', 'plant_organ2', 'collection_subset', 'collection_aka', 'county', 'protected', 'stratgroup', 'stratscale','reproduction','ontogeny', 'composition', 'architecture', 'thickness', 'collection_dates', 'collection_comments', 'taxon_environment', 'environment_basis', 'collectors'])
df[1]
```

```python
for i in df[1].columns:
    if df[1][str(i)].isna().sum() > (df[1].shape[0] * 3/4):
        del df[1][str(i)]
df[1].isna().sum()
```

```python
df[0]['name'] = df[0]['name'].str.capitalize()
df[1].rename(columns = {'accepted_name':'name'},inplace=True)
df[1]
```

```python
df[1]['diet'].unique()
```

```python
combined = pd.merge(df[1],df[0], on='name',how='outer')
combined.drop_duplicates().shape
```

```python
dino_df = df[1]
dino_df['diet'].dtype
```

```python
def classify_diet(diet):
    if 'carnivore' in diet:
        return 'carnivore'
    elif 'omnivore' in diet:
        return 'omnivore'
    else:
        return 'herbivore'

diet_class = []
for i in range(0,len(dino_df['diet'])):
    if pd.isna(dino_df['diet'].iloc[i]):
        diet_class.append(dino_df['diet'].iloc[i])
    else:
        diet_class.append(classify_diet(dino_df['diet'].iloc[i]))
dino_df['diet_class'] = diet_class
```

```python

pie_data = dino_df.groupby('early_interval').count().sort_values(by='occurrence_no',ascending = False).reset_index()[:10]
#plot = pie_data.plot.pie(y='occurrence_no', figsize=(10, 20))

sns.set_style("whitegrid") # Set style for chart
plt.figure(figsize=(6,6)) # Set figure size
plt.pie(pie_data['occurrence_no'], labels=pie_data['early_interval']) # Create pie chart
plt.show() # Show chart
```

```python
dino_df['mass_range'] = dino_df['max_ma'] - dino_df['min_ma']
```

```python
bar_data = dino_df[['early_interval','diet_class','mass_range']].groupby(['early_interval','diet_class']).mean()
bar_data
```

```python
test = bar_data.reset_index()
test = test[['early_interval','mass_range']].groupby(['early_interval']).mean().sort_values(by='mass_range',ascending=False)
era_name = test.index[:10]
era_name = list(era_name)
era_name
```

```python
plt.figure(figsize=(30,20))
bar_data = bar_data.reset_index()
bar_data = bar_data[bar_data['early_interval'].isin(era_name)]
sns.barplot(data = bar_data,x='early_interval',y='mass_range',hue='diet_class')
plt.xlabel('Era')
plt.ylabel('Average Mass')
plt.legend(Title = 'Diet')
plt.show()
```
